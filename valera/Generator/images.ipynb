```python
import tensorflow.keras 
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import os
from matplotlib import pyplot as plt
```

```python
from skimage.io import imread
```

```python
data_gen = ImageDataGenerator(
            featurewise_center=True,
            featurewise_std_normalization=True,
            rotation_range = 25,
            horizontal_flip = 1,
            width_shift_range=0.1,
            height_shift_range=0.1,
            channel_shift_range = 0.4,
            zoom_range = [0.8,1.5],
            fill_mode = "nearest",
)
```

```python
cat_images = []
dog_images = []
for path in os.listdir("C:\\Users\\PM65M\\OneDrive\\Desktop\\data\\input\\catmord\\"):
    k1 = imread("C:\\Users\\PM65M\\OneDrive\\Desktop\\data\\input\\catmord\\" + path)
    k = np.copy(k1).astype(np.float32)/ 255.
    h = data_gen.flow(np.array([k]))
    for j in range(20):
        imag_a = next(h)
        cat_images.append(imag_a)
        
for path in os.listdir("C:\\Users\\PM65M\\OneDrive\\Desktop\\data\\input\\dogmord\\"):
    k1 = imread("C:\\Users\\PM65M\\OneDrive\\Desktop\\data\\input\\dogmord\\" + path)
    k = np.copy(k1).astype(np.float32)/ 255.
    h = data_gen.flow(np.array([k]))
    for j in range(20):
        imag_a = next(h)
        dog_images.append(imag_a)
```

```python
import random
random.shuffle(cat_images)
random.shuffle(dog_images)
```

```python
from skimage.io import imsave
for i in range(1500):
    k = np.copy(cat_images[i][0] * 255).astype(np.int)
    imsave("C:\\Users\\PM65M\\OneDrive\\Desktop\\data\\output\\cat\\cat_v_" + str(i) + ".jpg",k)
    k = np.copy(dog_images[i][0] * 255).astype(np.int)
    imsave("C:\\Users\\PM65M\\OneDrive\\Desktop\\data\\output\\dog\\dog_v_" + str(i) + ".jpg",k)
```
